# Lecture - Built-In Judges

## Overview

MLflow provides research-validated built-in judges for common evaluation criteria like correctness, relevance, safety, and retrieval quality. This lecture introduces the spectrum of available judge types and takes a deep dive into built-in judges: the fastest way to start evaluating your agents.

## Learning Objectives

By the end of this lecture, you will be able to:
- Describe the spectrum of evaluation judge types available in MLflow
- Identify appropriate built-in judges for common evaluation criteria
- Understand which judges require ground truth and which do not
- Apply built-in judges using `mlflow.genai.evaluate()`

var EVAL_STEPS = [
  {
    title: "1. Define Your Agent",
    color: "#2272B4",
    bg: "rgba(34,114,180,0.10)",
    desc: "Your agent function should accept the keys from the <code>inputs</code> dictionary as keyword arguments and emit one trace per call. If your function doesn't already produce a trace, MLflow automatically applies <code>@mlflow.trace</code> for you. The return value can be any type (string, dict, list, etc.).",
    code: '@mlflow.trace\ndef my_agent(question: str) -> str:\n    """Your agent function."""\n    # Your agent logic here\n    return "..."'
  },
  {
    title: "2. Choose a Scorer",
    color: "#00A972",
    bg: "rgba(0,169,114,0.10)",
    desc: "Import and instantiate a built-in judge. You can optionally override the default judge model with the <code>model</code> parameter.",
    code: 'from mlflow.genai.scorers import Correctness\n\ncorrectness_eval = Correctness(\n    model=\"databricks:/foundation-model-endpoint\"\n)'
  },
  {
    title: "3. Build Evaluation Dataset",
    color: "#FF5F46",
    bg: "rgba(255,95,70,0.10)",
    desc: "Each record contains <code>inputs</code> (matching your agent's kwargs) and optionally <code>expectations</code> with ground truth facts or guidelines.",
    code: 'eval_data = [\n    {\"inputs\": {\"question\": \"What is MLflow?\"},\n     \"expectations\": {\"expected_facts\": [\"MLflow is an open-source platform\"]}},\n    {\"inputs\": {\"question\": \"What is Spark?\"},\n     \"expectations\": {\"expected_facts\": [\"Spark is a data engine\"]}},\n]'
  },
  {
    title: "4. Run Evaluation",
    color: "#1B5162",
    bg: "rgba(27,81,98,0.10)",
    desc: "Pass your dataset, agent function, and scorers to <code>mlflow.genai.evaluate()</code>. Results are logged to MLflow automatically.",
    code: 'results = mlflow.genai.evaluate(\n    data=eval_data,\n    predict_fn=my_agent,\n    scorers=[correctness_eval],\n)'
  }
];

var evalCurrent = null;

function selectEvalStep(id) {
  var wrap = document.getElementById('eval-detail-wrap');
  var card = document.getElementById('eval-detail-card');
  var s = EVAL_STEPS[id];

  document.querySelectorAll('.eval-step-box').forEach(function(b) {
    b.classList.toggle('active', parseInt(b.dataset.id, 10) === id);
  });

  if (evalCurrent === id) {
    wrap.classList.remove('open');
    document.querySelectorAll('.eval-step-box').forEach(function(b) { b.classList.remove('active'); });
    evalCurrent = null;
    return;
  }

  evalCurrent = id;
  card.style.borderTopColor = s.color;
  card.innerHTML =
    '<div style="font-size:17px;font-weight:700;margin-bottom:10px;color:#0b2026;">' + s.title + '</div>'
    + '<div style="font-size:14px;line-height:1.6;color:#333;margin-bottom:10px;">' + s.desc + '</div>'
    + '<div class="eval-code-block" style="background:' + s.bg + ';border-color:' + s.color + ';">' 
    + '<code>' + s.code.replace(/\n/g, '<br>') + '</code></div>';

  wrap.classList.add('open');
}

selectEvalStep(0);
</script>
