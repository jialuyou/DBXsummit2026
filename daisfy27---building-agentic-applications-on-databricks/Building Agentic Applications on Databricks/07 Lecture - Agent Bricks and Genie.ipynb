# Lecture - Agent Bricks and Genie
## Overview
In the previous notebooks you built agents with code using the OpenAI Agents SDK. Databricks also provides **Agent Bricks**, which is a no-code platform for building, optimizing, and governing production agents, and **Genie** for natural language access to structured data. This lecture covers both and discusses how a **Genie Agent** can be integrated with **Agent Bricks**. 

## Learning Objectives
By the end of this lecture, you will be able to:
1. Describe what Agent Bricks is and identify its supported agent types (Knowledge Assistant, Supervisor Agent, Document Intelligence, Custom Agents)
2. Explain how Agent Bricks differs from the code-first Agent Framework approach
3. Understand the Agent Bricks development lifecycle: specify, optimize, improve
4. Describe how Genie enables natural language querying of structured data
5. Identify when to use a Knowledge Assistant vs a Supervisor Agent vs Genie

## A. Agent Bricks
[Databricks Agent Bricks](https://www.databricks.com/de/product/artificial-intelligence/agent-bricks) is a unified platform and control plane for building, optimizing, and governing production AI agents over your enterprise data, shifting the focus from low-level implementation details to the actual business task, data, and quality metrics.

### A1. Core Three‑Step Development Cycle

The Agent Bricks development lifecycle is an iterative loop built around three core phases: **define the problem**, **configure and evaluate on your data**, and **continuously improve**. Each Agent Brick (Knowledge Assistant, Supervisor Agent, Classification, Information Extraction, and, where available, legacy Custom LLM flows) builds on the same underlying Databricks platform: Unity Catalog, Mosaic AI Model Serving, MLflow Tracing, and Agent Evaluation.

## B. Genie Agents

A **Genie space** is a curated natural-language interface over structured data registered in Unity Catalog (tables, views, metric views, etc.).

A **Genie agent** is simply Genie used as an agent on top of a specific Genie space (via the Genie APIs or managed MCP URL), so orchestrators can call it as a callable worker in multi-agent or app workflows.

## C. Conclusion

In this lecture, you learned about two production deployment paths on Databricks:

- **Agent Bricks**: a no-code platform for building Knowledge Assistants, Supervisor Agents, Document Intelligence, and Custom Agents, with automatic optimization and built-in evaluation
- **Genie**: natural language access to structured data via Genie Spaces, usable standalone or as a subagent within a Supervisor Agent

The choice between code-first (OpenAI SDK, LangGraph) and no-code (Agent Bricks) depends on your use case:
- Use **Agent Bricks** when you want managed optimization, built-in evaluation, and rapid deployment without writing agent code
- Use **code-first frameworks** when you need full control over orchestration logic, custom tool implementations, or complex multi-step workflows
